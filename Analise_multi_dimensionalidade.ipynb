## Como trabalhar com dados de alta dimensionalidade?

```python
# Importando pandas para iniciar as análises

import pandas as pd
```

```python
# Setando configurações básicas

pd.set_option('display.max_columns', 35)
pd.set_option('display.max_rows', 569)
```

```python
# Carregando base de dados

url = 'https://raw.githubusercontent.com/alura-cursos/reducao-dimensionalidade/master/data-set/exames.csv'
dados_exames = pd.read_csv(url)
dados_exames.head()
```

```python
# Dropando algumas features que não serão aproveitadas nas análises

dados_exames = dados_exames.drop(columns=['id', 'exame_33'])
dados_exames.head()
```

```python
# Importando lib para realizar o split dos dados e gerar aleatoriedade dos dados

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
# Separando as bases de dados para treino e teste

x = dados_exames.drop(columns=['diagnostico'])
y = dados_exames.diagnostico
```

```python
# Preparando a base

SEED = 123143
np.random.seed(SEED)

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, test_size=0.3)
```

```python
# Instanciando e treinando o modelo

classificador = RandomForestClassifier(n_estimators=100)
classificador.fit(treino_x, treino_y)
```

```python
# Gerando score do modelo

cf_score = classificador.score(teste_x, teste_y) * 100
print("Resultado da classificação é de %.2f%%" %cf_score)
```

#

## Criando Baseline

```python
# importando biblioteca para baseline

from sklearn.dummy import DummyClassifier
```

```python
# Setando classificador e gerando metricas

SEED = 123143
np.random.seed(SEED)

classificador_baseline = DummyClassifier(strategy="most_frequent")
classificador_baseline.fit(treino_x, treino_y)
print("Resultado do baseline é de: %.2f%%" %(classificador_baseline.score(teste_x, teste_y)*100))
```

### Plotando resultados para analisar o comportamento das features

```python
# Importando bibliotecas para analisar o comportamento das variáveis independentes

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Preparando a base para utilização do Seaborn

dados_plot = pd.concat([y, x.iloc[:,0:10]], axis=1) # Definindo a visualização dos 10 primeiros exames (iloc)
dados_plot = pd.melt(dados_plot, id_vars='diagnostico', var_name='exames', value_name='valores')
dados_plot.head()
```

```python
# Plotando visualização de violino

plt.figure(figsize=(10, 10))
sns.violinplot(x = 'exames',  y='valores', hue='diagnostico', data=dados_plot)
plt.xticks(rotation = 90)
plt.show()
```

```python
# Importando biblioteca de preprocessamento para padronizar o eixo x

from sklearn.preprocessing import StandardScaler

padronizador = StandardScaler()
padronizador.fit(x)
valores_exames = padronizador.transform(x)
valores_exames
valores_exames = pd.DataFrame(data = valores_exames, columns = x.keys())
valores_exames.head()
```

```python
# Preparando a base para nova utilização do Seaborn com dados padronizados

dados_plot = pd.concat([y, valores_exames], axis=1) # Definindo a visualização dos 10 primeiros exames (iloc)
dados_plot = pd.melt(dados_plot, id_vars='diagnostico', var_name='exames', value_name='valores')
dados_plot.head()
```

```python
# Plotando visualização de violino com dados padronizados

plt.figure(figsize=(10, 10))
sns.violinplot(x = 'exames',  y='valores', hue='diagnostico', data=dados_plot, split=True)
plt.xticks(rotation = 90)
plt.show()
```

```python
# Excluindo features que possuem valores constantes e não agregam ao modelo

valores_exames = valores_exames.drop(columns=['exame_4', 'exame_29'])
```

```python
# Criando função para avaliar o resultado dos dados atualizados

def classificar(valores):
    SEED = 1234
    np.random.seed(SEED)
    treino_x, teste_x, treino_y, teste_y = train_test_split(valores, y, test_size=0.3)
    classificador = RandomForestClassifier(n_estimators=100)
    classificador.fit(treino_x, treino_y)
    print("Resultado da classificação é de %.2f%%" %(classificador.score(teste_x, teste_y) * 100))    
```

```python
classificar(valores_exames)
```

### Feature Selection com chi2

```python
# importando lib de seleção das melhores features

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
# Instanciando modelo

selecionar_kmelhores = SelectKBest(chi2, k=5)
```

```python
# Utilizando base antes da normalização, pois o modelo do chi2 não permite utilização de valores negativos

valores_exames_sn = x.drop(columns=['exame_4', 'exame_29', 'exame_24', 'exame_3'], axis=1) 
```

```python
# Separando bases de treino e teste e realizando o treino do modelo

SEED = 1234
np.random.seed(SEED)
treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_sn, y, test_size=0.3)
selecionar_kmelhores.fit(treino_x, treino_y)
treino_kbest = selecionar_kmelhores.transform(treino_x)
teste_kbest = selecionar_kmelhores.transform(teste_x)
```

```python
# Rodando árvore de decisão com as 5 melhores features

classificador = RandomForestClassifier(n_estimators=100, random_state=1234)
classificador.fit(treino_kbest, treino_y)
print("Resultado da classificação é de %.2f%%" %(classificador.score(teste_kbest, teste_y) * 100))    
```

### Matriz de Confusão

```python
# Importando lib de matriz de confusão para avaliar as métricas

from sklearn.metrics import confusion_matrix
```

```python
# Instanciando matriz

matriz_confusao = confusion_matrix(teste_y, classificador.predict(teste_kbest))
```

```python
matriz_confusao
```

```python
# Plotando a matriz de confusão com lib gráfica

plt.figure(figsize=(10, 8))
sns.set(font_scale=2)
sns.heatmap(matriz_confusao, annot=True, fmt='d').set(xlabel='Predição', ylabel = 'Real')
plt.show()
```

### Aplicando RFE

```python
# Importando feature selection recursiva 

from sklearn.feature_selection import RFE

# Separando bases de treino e teste e realizando o treino do modelo

SEED = 1234
np.random.seed(SEED)
treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_sn, y, test_size=0.3)

# Rodando árvore de decisão com as 5 melhores features

classificador = RandomForestClassifier(n_estimators=100, random_state=1234)
classificador.fit(treino_x, treino_y)


# Instanciando modelo

sel_rfe = RFE(estimator= classificador, n_features_to_select=2 , step=1)

# treinando novo modelo

sel_rfe.fit(treino_x, treino_y)
treino_rfe = sel_rfe.transform(treino_x)
teste_rfe = sel_rfe.transform(teste_x)

# Treinando classificador

classificador.fit(treino_rfe, treino_y)

# Plotando a matriz de confusão com modelo rfe

matriz_confusao = confusion_matrix(teste_y, classificador.predict(teste_rfe))
plt.figure(figsize=(10, 8))
sns.set(font_scale=2)
sns.heatmap(matriz_confusao, annot=True, fmt='d').set(xlabel='Predição', ylabel = 'Real')
plt.show()

print("Resultado da classificação é de %.2f%%" %(classificador.score(teste_rfe, teste_y) * 100))    
```

#

### Utillizando o RFECV

```python
# Importando a biblioteca RFECV

from sklearn.feature_selection import RFECV

# Instanciando modelo
# Separando bases de treino e teste e realizando o treino do modelo

SEED = 1234
np.random.seed(SEED)
treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_sn, y, test_size=0.3)
classificador = RandomForestClassifier(n_estimators=100, random_state=1234)
classificador.fit(treino_x, treino_y)

sel_rfecv = RFECV(estimator= classificador, cv=5 , step=1, scoring='accuracy')

# treinando modelo RFECV

sel_rfecv.fit(treino_x, treino_y)
treino_rfecv = sel_rfecv.transform(treino_x)
teste_rfecv = sel_rfecv.transform(teste_x)

# Treinando classificador

classificador.fit(treino_rfecv, treino_y)

# Plotando a matriz de confusão com modelo rfecv

matriz_confusao = confusion_matrix(teste_y, classificador.predict(teste_rfecv))
plt.figure(figsize=(10, 8))
sns.set(font_scale=2)
sns.heatmap(matriz_confusao, annot=True, fmt='d').set(xlabel='Predição', ylabel = 'Real')
plt.show()

print("Resultado da classificação é de %.2f%%" %(classificador.score(teste_rfecv, teste_y) * 100))    
```

```python
# Plotando a importancia das features

plt.figure(figsize=(8,8))
plt.xlabel('Número de exames')
plt.ylabel('Acurácia')
plt.plot(range(1, len(sel_rfecv.grid_scores_) +1), sel_rfecv.grid_scores_)
plt.show() 
```
